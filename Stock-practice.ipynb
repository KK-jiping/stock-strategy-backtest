```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 三井不動産のデータ取得
ticker = "8801.T"
data = yf.download(ticker, start="2020-01-01", end="2024-12-31")["Close"]
data = data.squeeze()
data.name = "三井不動産"

# 移動平均の計算
short_window = 25
long_window = 75

df = pd.DataFrame(data)
df.columns = ["Close"]
df["MA25"] = df["Close"].rolling(window=short_window).mean()
df["MA75"] = df["Close"].rolling(window=long_window).mean()

# シグナルの生成
df["Signal"] = 0
df.loc[df["MA25"] > df["MA75"], "Signal"] = 1  # 買いポジション
df["Position"] = df["Signal"].diff()  # 売買タイミング

# 売買シグナルのプロット
plt.figure(figsize=(14, 7))
plt.plot(df.index, df["Close"], label="株価", alpha=0.7)
plt.plot(df.index, df["MA25"], label="25日移動平均", alpha=0.8)
plt.plot(df.index, df["MA75"], label="75日移動平均", alpha=0.8)

# 買いシグナル
buy_signals = df[df["Position"] == 1]
plt.scatter(buy_signals.index, buy_signals["Close"],
           marker="^", color="green", s=100, label="買いシグナル", zorder=5)

# 売りシグナル
sell_signals = df[df["Position"] == -1]
plt.scatter(sell_signals.index, sell_signals["Close"],
           marker="v", color="red", s=100, label="売りシグナル", zorder=5)

plt.title("三井不動産：移動平均クロス戦略（25日/75日）")
plt.xlabel("日付")
plt.ylabel("株価（円）")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# バックテスト
df["Daily_Return"] = df["Close"].pct_change()
df["Strategy_Return"] = df["Daily_Return"] * df["Signal"].shift(1)

# 累積リターンの計算
df["Cumulative_Market"] = (1 + df["Daily_Return"]).cumprod()
df["Cumulative_Strategy"] = (1 + df["Strategy_Return"]).cumprod()

# 累積リターンのプロット
plt.figure(figsize=(14, 6))
plt.plot(df.index, df["Cumulative_Market"],
         label="バイ＆ホールド（買って持ち続ける）", alpha=0.8)
plt.plot(df.index, df["Cumulative_Strategy"],
         label="移動平均クロス戦略", alpha=0.8)
plt.title("累積リターン比較：移動平均クロス戦略 vs バイ＆ホールド")
plt.xlabel("日付")
plt.ylabel("累積リターン（2020年1月=1）")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# パフォーマンス指標の表示
market_return = df["Cumulative_Market"].iloc[-1] - 1
strategy_return = df["Cumulative_Strategy"].iloc[-1] - 1
market_sharpe = df["Daily_Return"].mean() / df["Daily_Return"].std() * np.sqrt(252)
strategy_sharpe = df["Strategy_Return"].mean() / df["Strategy_Return"].std() * np.sqrt(252)

print(f"バイ＆ホールドの総リターン: {market_return:.2%}")
print(f"移動平均クロス戦略の総リターン: {strategy_return:.2%}")
print(f"バイ＆ホールドのシャープレシオ（年率）: {market_sharpe:.3f}")
print(f"移動平均クロス戦略のシャープレシオ（年率）: {strategy_sharpe:.3f}")
print(f"売買回数: {int(df['Position'].abs().sum())}")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 三井不動産のデータ取得
ticker = "8801.T"
data = yf.download(ticker, start="2020-01-01", end="2024-12-31")["Close"]
data = data.squeeze()

df = pd.DataFrame(data)
df.columns = ["Close"]

# 移動平均の計算
df["MA10"] = df["Close"].rolling(window=10).mean()
df["MA25"] = df["Close"].rolling(window=25).mean()
df["MA75"] = df["Close"].rolling(window=75).mean()
df["MA100"] = df["Close"].rolling(window=100).mean()

# 非対称シグナルの生成
df["Signal"] = 0
position = 0
signals = []

for i in range(len(df)):
    if pd.isna(df["MA100"].iloc[i]) or pd.isna(df["MA75"].iloc[i]):
        signals.append(0)
        continue

    # ポジションなし → 買いシグナル判定
    if position == 0:
        if df["MA10"].iloc[i] > df["MA100"].iloc[i]:
            position = 1
            signals.append(1)
        else:
            signals.append(0)

    # ポジションあり → 売りシグナル判定
    elif position == 1:
        if df["MA25"].iloc[i] > df["MA75"].iloc[i]:
            position = 0
            signals.append(-1)
        else:
            signals.append(1)

df["Signal"] = signals
df["Position"] = pd.Series(signals, index=df.index).replace({1: 1, -1: 0, 0: 0})

# 売買タイミングの特定
df["Trade"] = pd.Series(signals, index=df.index)

# プロット①：株価と移動平均・売買シグナル
plt.figure(figsize=(14, 7))
plt.plot(df.index, df["Close"], label="株価", alpha=0.7)
plt.plot(df.index, df["MA10"], label="MA10", alpha=0.6, linestyle="--")
plt.plot(df.index, df["MA100"], label="MA100", alpha=0.6, linestyle="--")
plt.plot(df.index, df["MA25"], label="MA25", alpha=0.6, linestyle=":")
plt.plot(df.index, df["MA75"], label="MA75", alpha=0.6, linestyle=":")

buy_signals = df[df["Trade"] == 1]
sell_signals = df[df["Trade"] == -1]

plt.scatter(buy_signals.index, buy_signals["Close"],
           marker="^", color="green", s=150, label="買いシグナル", zorder=5)
plt.scatter(sell_signals.index, sell_signals["Close"],
           marker="v", color="red", s=150, label="売りシグナル", zorder=5)

plt.title("三井不動産：非対称移動平均クロス戦略")
plt.xlabel("日付")
plt.ylabel("株価（円）")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# バックテスト
df["Daily_Return"] = df["Close"].pct_change()
df["Strategy_Return"] = df["Daily_Return"] * df["Position"].shift(1)

df["Cumulative_Market"] = (1 + df["Daily_Return"]).cumprod()
df["Cumulative_Strategy"] = (1 + df["Strategy_Return"]).cumprod()

# プロット②：累積リターン比較
plt.figure(figsize=(14, 6))
plt.plot(df.index, df["Cumulative_Market"],
         label="バイ＆ホールド", alpha=0.8)
plt.plot(df.index, df["Cumulative_Strategy"],
         label="非対称移動平均クロス戦略", alpha=0.8)
plt.title("累積リターン比較：非対称戦略 vs バイ＆ホールド")
plt.xlabel("日付")
plt.ylabel("累積リターン（2020年1月=1）")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# パフォーマンス指標
market_return = df["Cumulative_Market"].iloc[-1] - 1
strategy_return = df["Cumulative_Strategy"].iloc[-1] - 1
market_sharpe = df["Daily_Return"].mean() / df["Daily_Return"].std() * np.sqrt(252)
strategy_sharpe = (df["Strategy_Return"].mean() /
                   df["Strategy_Return"].std() * np.sqrt(252))
trade_count = int((df["Trade"].abs() == 1).sum() / 2)

print(f"バイ＆ホールドの総リターン:       {market_return:.2%}")
print(f"非対称戦略の総リターン:           {strategy_return:.2%}")
print(f"バイ＆ホールドのシャープレシオ:    {market_sharpe:.3f}")
print(f"非対称戦略のシャープレシオ:        {strategy_sharpe:.3f}")
print(f"売買回数:                         {trade_count}回")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# データ取得
ticker = "8801.T"
data = yf.download(ticker, start="2020-01-01", end="2024-12-31")["Close"]
data = data.squeeze()

df = pd.DataFrame(data)
df.columns = ["Close"]

# 移動平均の計算
df["MA10"] = df["Close"].rolling(window=10).mean()
df["MA25"] = df["Close"].rolling(window=25).mean()
df["MA75"] = df["Close"].rolling(window=75).mean()
df["MA100"] = df["Close"].rolling(window=100).mean()

# 乖離率の計算
df["Buy_Gap"] = (df["MA10"] - df["MA100"]) / df["MA100"]
df["Sell_Gap"] = (df["MA25"] - df["MA75"]) / df["MA75"]

# 非対称シグナルの生成（乖離率1%以上・最低保有期間20日）
position = 0
holding_days = 0
min_holding = 20
gap_threshold = 0.01
signals = []
trades = []

for i in range(len(df)):
    if pd.isna(df["MA100"].iloc[i]):
        signals.append(0)
        trades.append(0)
        continue

    if position == 1:
        holding_days += 1

    # ポジションなし → 買いシグナル判定
    if position == 0:
        if df["Buy_Gap"].iloc[i] > gap_threshold:
            position = 1
            holding_days = 0
            signals.append(1)
            trades.append(1)
        else:
            signals.append(0)
            trades.append(0)

    # ポジションあり・最低保有期間経過 → 売りシグナル判定
    elif position == 1 and holding_days >= min_holding:
        if df["Sell_Gap"].iloc[i] > gap_threshold:
            position = 0
            signals.append(0)
            trades.append(-1)
        else:
            signals.append(1)
            trades.append(0)

    # 最低保有期間未満はホールド
    else:
        signals.append(1)
        trades.append(0)

df["Signal"] = signals
df["Trade"] = trades

# プロット①：株価と売買シグナル
plt.figure(figsize=(14, 7))
plt.plot(df.index, df["Close"], label="株価", alpha=0.7)
plt.plot(df.index, df["MA10"], label="MA10", alpha=0.6, linestyle="--")
plt.plot(df.index, df["MA100"], label="MA100", alpha=0.6, linestyle="--")
plt.plot(df.index, df["MA25"], label="MA25", alpha=0.6, linestyle=":")
plt.plot(df.index, df["MA75"], label="MA75", alpha=0.6, linestyle=":")

buy_signals = df[df["Trade"] == 1]
sell_signals = df[df["Trade"] == -1]

plt.scatter(buy_signals.index, buy_signals["Close"],
           marker="^", color="green", s=150, label="買いシグナル", zorder=5)
plt.scatter(sell_signals.index, sell_signals["Close"],
           marker="v", color="red", s=150, label="売りシグナル", zorder=5)

plt.title("三井不動産：修正版非対称移動平均クロス戦略")
plt.xlabel("日付")
plt.ylabel("株価（円）")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# バックテスト
df["Daily_Return"] = df["Close"].pct_change()
df["Strategy_Return"] = df["Daily_Return"] * df["Signal"].shift(1)

df["Cumulative_Market"] = (1 + df["Daily_Return"]).cumprod()
df["Cumulative_Strategy"] = (1 + df["Strategy_Return"]).cumprod()

# プロット②：累積リターン比較
plt.figure(figsize=(14, 6))
plt.plot(df.index, df["Cumulative_Market"],
         label="バイ＆ホールド", alpha=0.8)
plt.plot(df.index, df["Cumulative_Strategy"],
         label="修正版非対称戦略", alpha=0.8)
plt.title("累積リターン比較：修正版非対称戦略 vs バイ＆ホールド")
plt.xlabel("日付")
plt.ylabel("累積リターン（2020年1月=1）")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# パフォーマンス指標
market_return = df["Cumulative_Market"].iloc[-1] - 1
strategy_return = df["Cumulative_Strategy"].iloc[-1] - 1
market_sharpe = (df["Daily_Return"].mean() /
                 df["Daily_Return"].std() * np.sqrt(252))
strategy_sharpe = (df["Strategy_Return"].mean() /
                   df["Strategy_Return"].std() * np.sqrt(252))
buy_count = (df["Trade"] == 1).sum()
sell_count = (df["Trade"] == -1).sum()

print(f"バイ＆ホールドの総リターン:       {market_return:.2%}")
print(f"修正版非対称戦略の総リターン:     {strategy_return:.2%}")
print(f"バイ＆ホールドのシャープレシオ:   {market_sharpe:.3f}")
print(f"修正版非対称戦略のシャープレシオ: {strategy_sharpe:.3f}")
print(f"買いシグナル回数:                 {buy_count}回")
print(f"売りシグナル回数:                 {sell_count}回")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# データ取得
ticker = "8801.T"
data = yf.download(ticker, start="2020-01-01", end="2024-12-31")["Close"]
data = data.squeeze()

df = pd.DataFrame(data)
df.columns = ["Close"]

# 移動平均の計算
df["MA10"] = df["Close"].rolling(window=10).mean()
df["MA25"] = df["Close"].rolling(window=25).mean()
df["MA75"] = df["Close"].rolling(window=75).mean()
df["MA100"] = df["Close"].rolling(window=100).mean()
df["MA200"] = df["Close"].rolling(window=200).mean()

# MACDの計算
df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean()
df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean()
df["MACD"] = df["EMA12"] - df["EMA26"]
df["MACD_Signal"] = df["MACD"].ewm(span=9, adjust=False).mean()
df["MACD_Hist"] = df["MACD"] - df["MACD_Signal"]

# 乖離率の計算
df["Buy_Gap"] = (df["MA10"] - df["MA100"]) / df["MA100"]
df["Sell_Gap"] = (df["MA25"] - df["MA75"]) / df["MA75"]

# トレンドフィルター付き非対称シグナルの生成
position = 0
holding_days = 0
min_holding = 20
gap_threshold = 0.01
signals = []
trades = []

for i in range(len(df)):
    if pd.isna(df["MA200"].iloc[i]):
        signals.append(0)
        trades.append(0)
        continue

    if position == 1:
        holding_days += 1

    # トレンドフィルター条件
    above_ma200 = df["Close"].iloc[i] > df["MA200"].iloc[i]
    macd_positive = df["MACD"].iloc[i] > 0

    # ポジションなし → 買いシグナル判定
    if position == 0:
        if (df["Buy_Gap"].iloc[i] > gap_threshold and
                above_ma200 and macd_positive):
            position = 1
            holding_days = 0
            signals.append(1)
            trades.append(1)
        else:
            signals.append(0)
            trades.append(0)

    # ポジションあり・最低保有期間経過 → 売りシグナル判定
    elif position == 1 and holding_days >= min_holding:
        if df["Sell_Gap"].iloc[i] > gap_threshold:
            position = 0
            signals.append(0)
            trades.append(-1)
        else:
            signals.append(1)
            trades.append(0)

    else:
        signals.append(1)
        trades.append(0)

df["Signal"] = signals
df["Trade"] = trades

# プロット①：株価・移動平均・売買シグナル
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10),
                                 sharex=True)

ax1.plot(df.index, df["Close"], label="株価", alpha=0.7)
ax1.plot(df.index, df["MA200"], label="MA200",
         alpha=0.8, linewidth=2, color="purple")
ax1.plot(df.index, df["MA10"], label="MA10",
         alpha=0.5, linestyle="--")
ax1.plot(df.index, df["MA100"], label="MA100",
         alpha=0.5, linestyle="--")

buy_signals = df[df["Trade"] == 1]
sell_signals = df[df["Trade"] == -1]

ax1.scatter(buy_signals.index, buy_signals["Close"],
           marker="^", color="green", s=150,
           label="買いシグナル", zorder=5)
ax1.scatter(sell_signals.index, sell_signals["Close"],
           marker="v", color="red", s=150,
           label="売りシグナル", zorder=5)

ax1.set_title("三井不動産：MACDフィルター付き非対称戦略")
ax1.set_ylabel("株価（円）")
ax1.legend()
ax1.grid(True)

# MACDのプロット
ax2.plot(df.index, df["MACD"], label="MACD", color="blue")
ax2.plot(df.index, df["MACD_Signal"],
         label="シグナル", color="orange")
ax2.bar(df.index, df["MACD_Hist"],
        label="ヒストグラム", color="gray", alpha=0.3)
ax2.axhline(y=0, color="black", linestyle="-", linewidth=0.5)
ax2.set_ylabel("MACD")
ax2.set_xlabel("日付")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()

# バックテスト
df["Daily_Return"] = df["Close"].pct_change()
df["Strategy_Return"] = df["Daily_Return"] * df["Signal"].shift(1)

df["Cumulative_Market"] = (1 + df["Daily_Return"]).cumprod()
df["Cumulative_Strategy"] = (1 + df["Strategy_Return"]).cumprod()

# プロット②：累積リターン比較
plt.figure(figsize=(14, 6))
plt.plot(df.index, df["Cumulative_Market"],
         label="バイ＆ホールド", alpha=0.8)
plt.plot(df.index, df["Cumulative_Strategy"],
         label="MACDフィルター付き非対称戦略", alpha=0.8)
plt.title("累積リターン比較：MACDフィルター付き戦略 vs バイ＆ホールド")
plt.xlabel("日付")
plt.ylabel("累積リターン（2020年1月=1）")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# パフォーマンス指標
market_return = df["Cumulative_Market"].iloc[-1] - 1
strategy_return = df["Cumulative_Strategy"].iloc[-1] - 1
market_sharpe = (df["Daily_Return"].mean() /
                 df["Daily_Return"].std() * np.sqrt(252))
strategy_sharpe = (df["Strategy_Return"].mean() /
                   df["Strategy_Return"].std() * np.sqrt(252))
buy_count = (df["Trade"] == 1).sum()
sell_count = (df["Trade"] == -1).sum()

print(f"バイ＆ホールドの総リターン:            {market_return:.2%}")
print(f"MACDフィルター付き戦略の総リターン:    {strategy_return:.2%}")
print(f"バイ＆ホールドのシャープレシオ:        {market_sharpe:.3f}")
print(f"MACDフィルター付き戦略のシャープレシオ:{strategy_sharpe:.3f}")
print(f"買いシグナル回数:                      {buy_count}回")
print(f"売りシグナル回数:                      {sell_count}回")
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# データ取得
ticker = "8801.T"
data = yf.download(ticker, start="2020-01-01", end="2024-12-31")["Close"]
data = data.squeeze()

df = pd.DataFrame(data)
df.columns = ["Close"]

# 移動平均の計算
df["MA10"] = df["Close"].rolling(window=10).mean()
df["MA25"] = df["Close"].rolling(window=25).mean()
df["MA75"] = df["Close"].rolling(window=75).mean()
df["MA100"] = df["Close"].rolling(window=100).mean()
df["MA200"] = df["Close"].rolling(window=200).mean()

# MACDの計算
df["EMA12"] = df["Close"].ewm(span=12, adjust=False).mean()
df["EMA26"] = df["Close"].ewm(span=26, adjust=False).mean()
df["MACD"] = df["EMA12"] - df["EMA26"]
df["MACD_Signal"] = df["MACD"].ewm(span=9, adjust=False).mean()
df["MACD_Hist"] = df["MACD"] - df["MACD_Signal"]

# 乖離率の計算
df["Buy_Gap"] = (df["MA10"] - df["MA100"]) / df["MA100"]
df["Sell_Gap"] = (df["MA25"] - df["MA75"]) / df["MA75"]

# シグナル生成（MACD強制売却条件追加）
position = 0
holding_days = 0
min_holding = 20
gap_threshold = 0.01
signals = []
trades = []

for i in range(len(df)):
    if pd.isna(df["MA200"].iloc[i]):
        signals.append(0)
        trades.append(0)
        continue

    if position == 1:
        holding_days += 1

    above_ma200 = df["Close"].iloc[i] > df["MA200"].iloc[i]
    macd_positive = df["MACD"].iloc[i] > 0

    # ポジションなし → 買いシグナル判定
    if position == 0:
        if (df["Buy_Gap"].iloc[i] > gap_threshold and
                above_ma200 and macd_positive):
            position = 1
            holding_days = 0
            signals.append(1)
            trades.append(1)
        else:
            signals.append(0)
            trades.append(0)

    # ポジションあり → 売りシグナル判定
    elif position == 1:
        # 条件①：MACDが0以下になったら強制売却
        if not macd_positive:
            position = 0
            holding_days = 0
            signals.append(0)
            trades.append(-1)

        # 条件②：最低保有期間経過後、売りシグナル発動
        elif holding_days >= min_holding and df["Sell_Gap"].iloc[i] > gap_threshold:
            position = 0
            holding_days = 0
            signals.append(0)
            trades.append(-1)

        # 条件③：ホールド継続
        else:
            signals.append(1)
            trades.append(0)

df["Signal"] = signals
df["Trade"] = trades

# プロット①：株価・移動平均・売買シグナル・MACD
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10), sharex=True)

ax1.plot(df.index, df["Close"], label="株価", alpha=0.7)
ax1.plot(df.index, df["MA200"], label="MA200",
         alpha=0.8, linewidth=2, color="purple")
ax1.plot(df.index, df["MA10"], label="MA10",
         alpha=0.5, linestyle="--")
ax1.plot(df.index, df["MA100"], label="MA100",
         alpha=0.5, linestyle="--")

buy_signals = df[df["Trade"] == 1]
sell_signals = df[df["Trade"] == -1]

ax1.scatter(buy_signals.index, buy_signals["Close"],
            marker="^", color="green", s=150,
            label="買いシグナル", zorder=5)
ax1.scatter(sell_signals.index, sell_signals["Close"],
            marker="v", color="red", s=150,
            label="売りシグナル", zorder=5)

ax1.set_title("三井不動産：MACD強制売却付き非対称戦略")
ax1.set_ylabel("株価（円）")
ax1.legend()
ax1.grid(True)

ax2.plot(df.index, df["MACD"], label="MACD", color="blue")
ax2.plot(df.index, df["MACD_Signal"], label="シグナル", color="orange")
ax2.bar(df.index, df["MACD_Hist"], label="ヒストグラム",
        color="gray", alpha=0.3)
ax2.axhline(y=0, color="black", linestyle="-", linewidth=0.5)
ax2.set_ylabel("MACD")
ax2.set_xlabel("日付")
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()

# バックテスト
df["Daily_Return"] = df["Close"].pct_change()
df["Strategy_Return"] = df["Daily_Return"] * df["Signal"].shift(1)

df["Cumulative_Market"] = (1 + df["Daily_Return"]).cumprod()
df["Cumulative_Strategy"] = (1 + df["Strategy_Return"]).cumprod()

# プロット②：累積リターン比較
plt.figure(figsize=(14, 6))
plt.plot(df.index, df["Cumulative_Market"],
         label="バイ＆ホールド", alpha=0.8)
plt.plot(df.index, df["Cumulative_Strategy"],
         label="MACD強制売却付き戦略", alpha=0.8)
plt.title("累積リターン比較：MACD強制売却付き戦略 vs バイ＆ホールド")
plt.xlabel("日付")
plt.ylabel("累積リターン（2020年1月=1）")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# パフォーマンス指標
market_return = df["Cumulative_Market"].iloc[-1] - 1
strategy_return = df["Cumulative_Strategy"].iloc[-1] - 1
market_sharpe = (df["Daily_Return"].mean() /
                 df["Daily_Return"].std() * np.sqrt(252))
strategy_sharpe = (df["Strategy_Return"].mean() /
                   df["Strategy_Return"].std() * np.sqrt(252))
buy_count = (df["Trade"] == 1).sum()
sell_count = (df["Trade"] == -1).sum()

# 最大ドローダウンの計算
rolling_max = df["Cumulative_Strategy"].cummax()
drawdown = (df["Cumulative_Strategy"] - rolling_max) / rolling_max
max_drawdown_strategy = drawdown.min()

rolling_max_market = df["Cumulative_Market"].cummax()
drawdown_market = (df["Cumulative_Market"] - rolling_max_market) / rolling_max_market
max_drawdown_market = drawdown_market.min()

print(f"バイ＆ホールドの総リターン:              {market_return:.2%}")
print(f"MACD強制売却付き戦略の総リターン:        {strategy_return:.2%}")
print(f"バイ＆ホールドのシャープレシオ:          {market_sharpe:.3f}")
print(f"MACD強制売却付き戦略のシャープレシオ:    {strategy_sharpe:.3f}")
print(f"バイ＆ホールドの最大ドローダウン:        {max_drawdown_market:.2%}")
print(f"MACD強制売却付き戦略の最大ドローダウン:  {max_drawdown_strategy:.2%}")
print(f"買いシグナル回数:                        {buy_count}回")
print(f"売りシグナル回数:                        {sell_count}回")
```
